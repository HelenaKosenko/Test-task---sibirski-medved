```python
df_m.info()
```

```python
df_m.columns
```

```python
print(' :', df_m['/_'].min())
print(' :', df_m['/_'].max())
```

```python
df_m = df_m[df_m['/_'] > '2023-05-01 00:00:00']
df_m = df_m[df_m['/_'] < '2023-06-01 00:00:00']
```
